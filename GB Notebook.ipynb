# Monthly Sales Overview: Using Pandas, create a data frame that summarizes the total sales amount for each month.

```python
# Extract month and year from the 'invoice_date' column
df['invoice_month'] = df['invoice_date'].dt.month
df['invoice_year'] = df['invoice_date'].dt.year

# Create a new DataFrame to store monthly sales overview
monthly_sales = df.groupby(['invoice_year', 'invoice_month'])['invoice_amount'].sum().reset_index()

# Display the monthly sales overview
print(monthly_sales)
```

```python
# Set figure size
plt.figure(figsize=(12, 8))

# Create a bar plot for monthly sales
sns.barplot(x='invoice_month', y='invoice_amount', hue='invoice_year', data=monthly_sales, palette='viridis')

# Set plot labels and title
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.title('Monthly Sales Overview')

# Show the legend
plt.legend(title='Year')

# Show the plot
plt.show()
```

# Which product brands contribute the most to the total revenue, and how has this changed over the years?

```python
# Calculate total revenue for each brand and year
revenue_by_brand_year = df.groupby(['invoice_year', 'Brand'])['invoice_amount'].sum().reset_index()

# Identify the top-contributing brands for each year
top_brands_by_year = revenue_by_brand_year.loc[revenue_by_brand_year.groupby('invoice_year')['invoice_amount'].idxmax()]

# Set a larger figure size
plt.figure(figsize=(14, 8))

# Create a bar chart to show how the top-contributing brands change over the years
sns.barplot(x='invoice_year', y='invoice_amount', hue='Brand', data=top_brands_by_year, palette='viridis')
plt.title('Top-Contributing Brands to Total Revenue Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside the plot

# Show the plot
plt.show()
```

# Can we identify any outliers in the distribution of invoice amounts, and how do they impact the overall analysis?

```python
plt.figure(figsize=(12, 8))

# Create a boxplot to visualize the distribution of invoice amounts
sns.boxplot(x='invoice_amount', data=df, palette='Set3')
plt.title('Distribution of Invoice Amounts')
plt.xlabel('Invoice Amount')

# Show the plot
plt.show()

# Calculate z-scores to identify outliers
z_scores = (df['invoice_amount'] - df['invoice_amount'].mean()) / df['invoice_amount'].std()

# Set a threshold for identifying outliers (e.g., z-score > 3 or < -3)
outliers = df[abs(z_scores) > 3]

# Display the rows containing outliers
print("Rows with outliers:")
print(outliers)
```
